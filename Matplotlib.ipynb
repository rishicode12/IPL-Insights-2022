# Matplotlib - 3

```python
import matplotlib.pyplot as plt
```

# Linear Graph

```python
x =[1,2,3,4]
y =[5,6,7,8]

plt.plot(x,y)
plt.show()
```

```python
x =[1,2,3,4]
y =[5,6,7,8]
c = 'g'
plt.plot(x,y,c)
plt.show()
```

# Non-Linear Graph

```python
x =[5,7,9,4,6]
y =[12,3,3,9,1]

plt.plot(x,y)
plt.show()
```

```python
x =[5,7,9,4,6]
y =[12,3,3,9,1]

plt.plot(x,y, marker='o') # here marker help to make . in graph
plt.show()
```

```python
x =[5,7,9,4,6]
y =[12,3,3,9,1]

plt.plot(x,y)
plt.plot(y, marker= ">")
plt.plot(x, marker= "o")
plt.show()
```

# Bar Graph

```python
x= [1,2,3,4]
y= [5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x= [1,2,3,4]
y= [5,6,7,8]
c= ['r','g','b','y']
plt.bar(x,y,color= c)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5,6]
y =[2,4,5,6,8,9]
plt.title("Scatter plot")
plt.xlabel("Month")
plt.ylabel("Values")  

c =['red', 'magenta', 'green', 'skyblue', 'orange', 'blue']
plt.scatter(x,y, color= c, s=90)  #s = size
plt.show()
```

# PieChart

```python
import matplotlib.pyplot as plt
```

```python
x= [10,20,40,30]
y= ['english','math','social science','technology']

plt.pie(x, labels = y)
plt.show()
```

```python
x= [10,20,40,30]
y= ['english','math','social science','technology']
c= ['yellow','lightgreen','blue','brown']
plt.pie(x, labels = y, colors=c)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r"flowers.jpg"

img = Image.open(fname).convert('L')

plt.figure(figsize=(6,6))     # important
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.savefig("output1.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python
x= [10,20,40,30]
y= ['english','math','social science','hindi']

plt.pie(x, labels = y)
plt.legend()
plt.show()
```
